# Description

This code imports the dataframes with the equilibrium and time resolved optical properties of Ta2NiSe5 and analyses the effect of changing pump fluence and varying temperature. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Import dataframes

```python
RT150_S1fit = pd.read_csv('../data/Dataframes/RT150_S1.csv')
RT900_S1fit = pd.read_csv('../data/Dataframes/RT900_S1.csv')
RT3500_S1fit = pd.read_csv('../data/Dataframes/RT3500_S1.csv')
RT4000_S1fit = pd.read_csv('../data/Dataframes/RT4000_S1.csv')
RT5000_S1fit = pd.read_csv('../data/Dataframes/RT5000_S1.csv')
RT6000_S1fit = pd.read_csv('../data/Dataframes/RT6000_S1.csv')

HT250_S1fit = pd.read_csv('../data/Dataframes/HT250_S1.csv')
HT3000_S1fit = pd.read_csv('../data/Dataframes/HT3000_S1.csv')
```

### Pick parameters to plot

Times stamps are not the same for all files so we need to use indices

```python
Xcols = 3 #number of columns for the x axis
xlimwide = [0.87,1.7]
xlimnarrow = [0.87,1.15]
```

### Comparison between different fluences at room temperature

```python
Times = Xcols + np.array([0,3,5,10,18,RT150_S1fit.shape[1]-Xcols-1])

fig, axes = plt.subplots(2,4,figsize=(25,14))

RT150_S1fit.plot(x="En",y=RT150_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,3550), ax=axes[0,0],
                 title = "150 uW")
RT900_S1fit.plot(x="En",y=RT900_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,3650), ax=axes[0,1],
                 title = "900 uW")
RT5000_S1fit.plot(x="En",y=RT5000_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,4250), ax=axes[0,2],
                 title = "5000 uW")
RT6000_S1fit.plot(x="En",y=RT6000_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,4250), ax=axes[0,3],
                 title = "6000 uW")
RT150_S1fit.plot(x="En",y=RT150_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2420,2650), ax=axes[1,0],
                 title = "150 uW")
RT900_S1fit.plot(x="En",y=RT900_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2300,2650), ax=axes[1,1],
                 title = "900 uW")
RT5000_S1fit.plot(x="En",y=RT5000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2350,3050), ax=axes[1,2],
                 title = "5000 uW")
RT6000_S1fit.plot(x="En",y=RT6000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2350,3050), ax=axes[1,3],
                 title = "6000 uW")

for ax in axes.flat:
    ax.set_xlabel('Energy (eV)',fontsize=14)
    ax.set_ylabel('$\sigma_{1}$ ($\Omega^{-1}$ cm$^{-1}$)',fontsize=14)
```

```python
Times = Xcols + np.array([0,3,5,10,18,RT150_S1fit.shape[1]-Xcols-1])

fig, axes = plt.subplots(1,2,figsize=(10,6))

RT150_S1fit.plot(x="En",y=RT150_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,3550), ax=axes[0],
                 title = "26 $\mu$J/cm$^{2}$",fontsize=14)
RT150_S1fit.plot(x="En",y=RT150_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2420,2650), ax=axes[1],
                 title = "26 $\mu$J/cm$^{2}$",fontsize=14)

for ax in axes.flat:
    ax.set_xlabel('Energy (eV)',fontsize=14)
    ax.set_ylabel('$\sigma_{1}$ ($\Omega^{-1}$ cm$^{-1}$)',fontsize=14)
    
plt.tight_layout()
```

### Comparison between low and high temperature

```python
Times = Xcols + np.array([0,1,2,3,RT150_S1fit.shape[1]-Xcols-1])

fig, axes = plt.subplots(2,4,figsize=(25,14))

RT150_S1fit.plot(x="En",y=RT150_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,3550), ax=axes[0,0],
                 title = "RT, 150 uW")
RT5000_S1fit.plot(x="En",y=RT5000_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2350,4250), ax=axes[0,1],
                 title = "RT, 5000 uW")
HT250_S1fit.plot(x="En",y=HT250_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2400,3650), ax=axes[0,2],
                title = "HT, 250 uW")
HT3000_S1fit.plot(x="En",y=HT3000_S1fit.iloc[:,Times].columns,xlim=xlimwide,ylim=(2400,3650), ax=axes[0,3],
                title = "HT, 3000 uW")
RT150_S1fit.plot(x="En",y=RT150_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2420,2650), ax=axes[1,0],
                 title = "RT, 150 uW")
RT5000_S1fit.plot(x="En",y=RT5000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2350,2650), ax=axes[1,1],
                 title = "RT, 5000 uW")
HT250_S1fit.plot(x="En",y=HT250_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2400,2500), ax=axes[1,2],
                title = "HT, 250 uW")
HT3000_S1fit.plot(x="En",y=HT3000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2400,2650), ax=axes[1,3],
                title = "HT, 3000 uW")

for ax in axes.flat:
    ax.set_xlabel('Energy (eV)',fontsize=14)
    ax.set_ylabel('$\sigma_{1}$ ($\Omega^{-1}$ cm$^{-1}$)',fontsize=14)
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,8))

Times = Xcols + np.array([0,1,6,9,RT150_S1fit.shape[1]-Xcols-1])

HT3000_S1fit.plot(x="En",y=HT3000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2420,2650), ax=axes[0],
                title = "343K, 176 $\mu$J/cm$^{2}$",fontsize=14)

Times = Xcols + np.array([0,1,2,3]+list(range(7,8))+[11])
RT4000_S1fit.plot(x="En",y=RT4000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2420,2650), ax=axes[1],
                 title = "293 K, 705 $\mu$J/cm$^{2}$, short delays",fontsize=14)

Times = Xcols + np.array([0]+[11,15,16,19,24])
RT4000_S1fit.plot(x="En",y=RT4000_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=(2420,2650), ax=axes[2],
                 title = "293 K, 705 $\mu$J/cm$^{2}$, long times",fontsize=14)

for ax in axes.flat:
    ax.set_xlabel('Energy (eV)',fontsize=14)
    ax.set_ylabel('$\sigma_{1}$ ($\Omega^{-1}$ cm$^{-1}$)',fontsize=14)
    
plt.tight_layout()
```

### Time evolution, RT, 900 uW

```python
ylim = [2350,2750]

fig, axes = plt.subplots(1,2,figsize=(12,8))

Times = Xcols + np.array([0,1,2])
RT900_S1fit.plot(x="En",y=RT900_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=ylim, ax=axes[0],
                 title = "RT, 900 uW, short times")

Times = Xcols + np.array([0]+[10,20])
RT900_S1fit.plot(x="En",y=RT900_S1fit.iloc[:,Times].columns,xlim=xlimnarrow,ylim=ylim, ax=axes[1],
                 title = "RT, 900 uW, long times")


for ax in axes.flat:
    ax.set_xlabel('Energy (eV)',fontsize=14)
    ax.set_ylabel('$\sigma_{1}$ ($\Omega^{-1}$ cm$^{-1}$)',fontsize=14)
```
